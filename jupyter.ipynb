```python
import os.path
import sys
from pathlib import Path
import random

import streamlit as st
import pandas as pd
import json
import csv

CURRENT_DIR = Path().absolute()
sys.path.append('./data')

from style_labels import style_400_keys

ESSENTIA_ANALYSIS_PATH = 'data/extended_descriptors_output_processed.json'

# Read the JSON data
with open(ESSENTIA_ANALYSIS_PATH, 'r') as f:
    data = json.load(f)
    df = pd.DataFrame(data)

df.head()
```

```python
# Define a function to extract the style activation values from the JSON data
def extract_style_activations(data):
    activations = []
    for item in data:
        activation = {}
        # Convert the style activations from JSON string to dictionary
        style_activations_dict = json.loads(item['style_activations'])
        for key, value in style_activations_dict.items():
            activation[key] = value
        activations.append(activation)
    return activations
# Extract the style activations from the JSON data
style_activations = extract_style_activations(data)

# Create a new dataframe with 400 columns for each style tag
style_df = pd.DataFrame(columns=style_400_keys)

# Fill in the values for each style tag column based on the style activations
for i, activation in enumerate(style_activations):
    for key, value in activation.items():
        style_df.loc[i, key] = value

# Combine the original dataframe with the new style dataframe
df_with_styles = pd.concat([df, style_df], axis=1)
```
